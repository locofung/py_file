# merge

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5'],
                   'age':[10,20,30,40,50,60]
                   })  
df1
```

```python
df2 = pd.DataFrame({'keys': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2'],
                      'dept':['IT','HR','REMIT']
                })
df2
```

```python
df4=pd.merge(df1,df2, how='left', left_on='key', right_on='keys')[['A','B','key']]
df4
```

```python
df5=pd.merge(df1,df2, how='inner', left_on='key', right_on='keys')[['A','B','key']]
df5
```

```python
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
technologies3 = {
    'Instructor': ["Jerry", "Tom", "Smith", "John"],
    'Courses': ["Spark", "Java", "Python", "Scala"]
}
```

```python
dft1=pd.DataFrame(technologies)
dft2=pd.DataFrame(technologies2)
dft3=pd.DataFrame(technologies3)
dft2
```

```python
df4=dft1.merge(dft2,on='Courses').merge(dft3,on='Courses')[['Courses','Fee','Instructor']]
df4
```
